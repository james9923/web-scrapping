```python
import requests
```

```python
from bs4 import BeautifulSoup as bs
```

```python
import csv
```

```python
data = requests.get('https://www.jumia.com.ng/air-conditioners-accessories/')
```

```python
print(data)
```

```python
print(data.content)
```

```python
result = bs(data.content, 'html.parser')
```

```python
print(result)
```

```python
print(result.prettify())
```

```python
r1 = result.find_all('h3', class_="name")
print(r1)
```

```python
for x in r1:
    print(x.text)
```

```python
r2 = result.find_all('div', class_="prc")
print(r2)
```

```python
print(len(r2))
```

```python
print(len(r1))
```

```python
for price in r2:
    print(price.text)
```

```python
products = result.find_all('div', class_ = "info")
```

```python
for name in products:
    product_name = name.h3.text
    product_price = name.div.text
    
    print(product_name + "---------"+ product_price)
```

```python
r3 = result.find_all('div', class_="stars _s")
print(r3)
```

```python
for rating in r3:
    print(rating.text)
```

```python
print(len(r3))
```

```python
print(products)
```

```python
for l in products:
    print(l.div(class_="stars _s"))
```

```python
header = ['Product Name', 'Price']

with open("JumiaDataSet.csv", 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    
    for items in products:
        product_name = items.h3.text
        product_price = items.div.text
        d1 = [product_name, product_price]
        writer.writerow(d1)
```
