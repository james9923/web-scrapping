```python
ji = requests.get('https://jiji.ng/22-air-conditioner/hisense')
```

```python
jiji = bs(ji.content, 'html.parser')
```

```python
print(jiji.prettify())
```

```python
p_name = jiji.find_all('h4',class_="qa-advert-title b-list-advert__item-title")
len(p_name)
```

```python
p_price = jiji.find_all('p', class_="b-list-advert__item-price")
len(p_price)
```

```python
data_list = []

temp_dict = {}

#temp_dict['Product Name'] = p_name
#temp_dict ['Price'] = p_price
#data_list.append(temp_dict)

for pro_name in p_name:
    temp_dict['Product Name'] = pro_name.text
    data_list.append(temp_dict)
    
for pro_price in p_price:
    temp_dict['Price'] = pro_price.text
    data_list.append(temp_dict)
```
