```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
data = requests.get('https://www.jumia.com.ng/air-conditioners-accessories/')
```

```python
print(data)
```

```python
result = BeautifulSoup(data.content, 'html.parser')
```

```python
print(result.prettify())
```

```python
r1 = result.find_all('h3', class_="name")
```

```python
print(r1)
```

```python
for x in r1:
    i = x.text
    print(i)
```

```python
r2 = result.find_all('div', class_="prc")
print(r2)
```

```python
for price in r2:
    print(price.text)
```

```python
products = result.find_all('div', class_ = "info")
```

```python
for name in products:
    product_name = name.h3.text
    product_price = name.div.text
    
    print(product_name + "---------"+ product_price)
```
