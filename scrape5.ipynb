```python
import requests
```

```python
from bs4 import BeautifulSoup as bs
```

```python
import csv
```

```python
df = requests.get('https://www.thecodecenter.org/catalogue.html')
```

```python
print(df)
```

```python
data = bs(df.content, 'html.parser')
```

```python
print(data.prettify())
```

```python
print(data.title)
```

```python
print(data.title.text)
```

```python
divs = data.find_all('div')
```

```python
print(divs)
```

```python
h4 = data.find_all('h4')
```

```python
print(h4)
```

```python
for x in h4:
    print(x.text)
```

```python
info = data.find_all('div', class_="card-body")
print(info)
```

```python
for m in info:
    print(m.p.text)
```

```python
print(data.find_all('ul'))
```

```python
infomate = data.find_all('ul')
```

```python
for z in infomate:
    print(z.li.text)
```

```python
content = data.find_all('li')
for c in content:
    print(c.text)
```
